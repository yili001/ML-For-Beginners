# Nigerian Music scraped from Spotify - an analysis

```python
import pandas as pd

csv = pd.read_csv("../data/nigerian-songs.csv")
csv.head(5)
```

```python
# check the count of null cell.
csv.info()
# double check the count of null cell.
csv.isnull().sum() # no null cell
```

```python
#show the statistics of data like count, mean, 25% mean of every column.
csv.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#get the sum of different artist_top_genre.
genre_counts = csv["artist_top_genre"].value_counts()
#show the plot
plt.figure(figsize=(10,7))
# show top 10.
sns.barplot(x=genre_counts[:10].index,y=genre_counts[:10].values)
plt.xticks(rotation=45)
plt.xlabel(xlabel="artist_top_genre",fontdict={"color": "yellow"})
plt.title('Top genres',color = 'blue')
```

```python
# get all rows of containing "Missing" in column "artist_top_genre"
dropped_csv = csv[csv["artist_top_genre"] == "Missing"]
# the third group : "Missing", means cell data is missing,
# drop all cell of of this value in artist_top_genre column.
df = csv[csv["artist_top_genre"] != "Missing"]
```

```python
# show the correlation between columns.
correlation = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(correlation, vmax=.8, square=True)
# from the plot we can tell that only "energy" and "loudness"
# have strong correlation.
```

```python
# focus on artist_top_genre in "fro dancehall", "afropop", "nigerian pop" top 3.
df = df[(df['artist_top_genre'] == 'afro dancehall') | (df['artist_top_genre'] == 'afropop') | (df['artist_top_genre'] == 'nigerian pop')]
# drop rows which popularity <= 0
df = df[(df['popularity'] > 0)]
# now redraw the top 3 plot to show the count.
top = df['artist_top_genre'].value_counts()
plt.figure(figsize=(10,7))
sns.barplot(x=top.index,y=top.values)
plt.xticks(rotation=45)
plt.title('Top genres',color = 'blue')
```

```python
sns.set_theme(style="ticks")
# examine top three genres data distribution for
# popularity and danceability
g = sns.jointplot(
    data=df,
    x="popularity", y="danceability", hue="artist_top_genre",
    kind="kde",
)
```

```python
# show popularity and danceability relationship with three genres type distinguished.
sns.FacetGrid(df, hue="artist_top_genre", size=5) \
   .map(plt.scatter, "popularity", "danceability") \
   .add_legend()
```

```python
# show popularity and danceability relationship without three genres type distinguished.
sns.scatterplot(x=df["popularity"],y=df["danceability"])
```
